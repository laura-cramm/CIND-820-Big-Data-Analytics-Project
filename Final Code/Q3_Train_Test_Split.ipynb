```python
import numpy as np
import pandas as pd
```

```python
question3 = pd.read_pickle("brfss.pickle")
```

```python
#Removing rows that are not labeled
question3 = question3.loc[(question3["DECIDE"].isna() == False), :]
```

```python
#Checking that unlabeled rows were succesfully removed
question3['DECIDE'].isna().sum()
```

The BRFSS dataset was split into two using the train_test_split function of the sklearn library. 80% of the data was allocated into the training and validation portion, while 20% was allocated to the testing portion. This split is essential to allow the machine learning models to be tested on data the models were not trained on; this is true to a real-life scenario. Testing a model on data that it was trained on would result in an inaccurately optimistic evaluation of the model's performance.

```python
#Training/Validation & Test Split
from sklearn.model_selection import train_test_split

#Make one dataset without outcome, one with only outcome
q3features = question3.loc[:,  question3.columns != 'DECIDE']
q3target = question3['DECIDE']



q3features_trainvalid, q3features_test, q3target_trainvalid, q3target_test = train_test_split (q3features, q3target, random_state = 123,
                                test_size = 0.20, shuffle = True)
```

```python
q3features_trainvalid.to_pickle('q3features_trainvalid.pickle')
q3features_test.to_pickle('q3features_test.pickle')
q3target_trainvalid.to_pickle('q3target_trainvalid.pickle')
q3target_test.to_pickle('q3target_test.pickle')
```
