Logistic Regression

```python
import numpy as np
import pandas as pd
import statistics
from scipy import stats
```

```python
logRegScores1 = pd.read_csv("/content/logRegScores1.csv")
logRegScores2 = pd.read_csv("/content/logRegScores2.csv")
logRegScores3 = pd.read_csv("/content/logRegScores3.csv")
logRegScores4 = pd.read_csv("/content/logRegScores4.csv")
logRegScores5 = pd.read_csv("/content/logRegScores5.csv")
```

```python
logRegScores1 = logRegScores1.drop(['Unnamed: 0'], axis=1)
logRegScores2 = logRegScores2.drop(['Unnamed: 0'], axis=1)
logRegScores3 = logRegScores3.drop(['Unnamed: 0'], axis=1)
logRegScores4 = logRegScores4.drop(['Unnamed: 0'], axis=1)
logRegScores5 = logRegScores5.drop(['Unnamed: 0'], axis=1)
```

```python
merged_logRegA = logRegScores1.merge(logRegScores2,on=['Threshold'], suffixes=('_x1', '_x2'))
merged_logRegB = merged_logRegA.merge(logRegScores3,on=['Threshold'])
merged_logRegC = merged_logRegB.merge(logRegScores4,on=['Threshold'], suffixes=('_x3', '_x4'))
merged_logReg = merged_logRegC.merge(logRegScores5,on=['Threshold'])
merged_logReg.rename(columns={'F2_Score': 'F2_Score_x5', 'Accuracy': 'Accuracy_x5', 'Recall': 'Recall_x5', 'Precision': 'Precision_x5'}, inplace=True)
```

```python
merged_logReg= merged_logReg.loc[~( (merged_logReg['Recall_x1'] == 0) | (merged_logReg['Recall_x2'] ==0)| (merged_logReg['Recall_x3'] ==0)| (merged_logReg['Recall_x4'] ==0)| (merged_logReg['Recall_x5'] ==0)| (merged_logReg['Precision_x1'] ==0)| (merged_logReg['Precision_x2'] ==0)| (merged_logReg['Precision_x3'] ==0)| (merged_logReg['Precision_x4'] ==0)| (merged_logReg['Precision_x5'] ==0)      ),:]
```

1. Optimizing F2-Score

```python
#Kruskal-Wallis test
f2List=[]
for i in range(len(merged_logReg['F2_Score_x1'])):
  f2List.append([merged_logReg['F2_Score_x1'].values[i], merged_logReg['F2_Score_x2'].values[i],  merged_logReg['F2_Score_x3'].values[i],  merged_logReg['F2_Score_x4'].values[i],  merged_logReg['F2_Score_x5'].values[i]])

result = stats.kruskal(*f2List)
print(result)
```

```python
#There is a statistically significant difference in the population medians
```

```python
#Computing median F-2 Score across all five sliding windows for each combination
#of hyperparameters
for i in range(len(merged_logReg['F2_Score_x1'])):
      merged_logReg.loc[merged_logReg.index[i], 'Median_Score'] = statistics.median([merged_logReg.loc[merged_logReg.index[i], 'F2_Score_x1'],merged_logReg.loc[merged_logReg.index[i], 'F2_Score_x2'], merged_logReg.loc[merged_logReg.index[i], 'F2_Score_x3'], merged_logReg.loc[merged_logReg.index[i], 'F2_Score_x4'], merged_logReg.loc[merged_logReg.index[i], 'F2_Score_x5']] )

merged_logReg.sort_values("Median_Score", ascending=False, inplace=True)
merged_logReg

#Use threshold=0.1
```

2. Optimizing Accuracy

```python
#Kruskal-Wallis test
accuracyList=[]
for i in range(len(merged_logReg['Accuracy_x1'])):
  accuracyList.append([merged_logReg['Accuracy_x1'].values[i], merged_logReg['Accuracy_x2'].values[i],  merged_logReg['Accuracy_x3'].values[i],  merged_logReg['Accuracy_x4'].values[i],  merged_logReg['Accuracy_x5'].values[i]])

result = stats.kruskal(*accuracyList)
print(result)
```

```python
#There is a statistically significant difference in the population medians
```

```python
#Computing median accuracy across all five sliding windows for each combination
#of hyperparameters
for i in range(len(merged_logReg['Accuracy_x1'])):
      merged_logReg.loc[merged_logReg.index[i], 'Median_Accuracy'] = statistics.median([merged_logReg.loc[merged_logReg.index[i], 'Accuracy_x1'],merged_logReg.loc[merged_logReg.index[i], 'Accuracy_x2'], merged_logReg.loc[merged_logReg.index[i], 'Accuracy_x3'], merged_logReg.loc[merged_logReg.index[i], 'Accuracy_x4'], merged_logReg.loc[merged_logReg.index[i], 'Accuracy_x5']] )

merged_logReg.sort_values("Median_Accuracy", ascending=False, inplace=True)
merged_logReg

#Use threshold of 0.5
```
